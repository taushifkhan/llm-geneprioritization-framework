```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
gs_data_stat = pd.read_csv("./data/goldStandard_genewise_scores.csv",index_col=0)
gs_data_stat.head()
```

```python
plt.figure(figsize=(10, 4))
sns.lineplot(data=gs_data_stat, x=gs_data_stat.index, y='naive_wt_score', label='NaiveLLM', marker='o', color="#1f77b4")
sns.lineplot(data=gs_data_stat, x=gs_data_stat.index, y='rag_wt_score', label='RAG', marker='o', color='#17becf', alpha=0.7)
sns.lineplot(data=gs_data_stat, x=gs_data_stat.index, y='hybrid_wt_score', label='HybridLLM', marker='s',color="#2ca02c",alpha=0.6)
sns.lineplot(data=gs_data_stat, x=gs_data_stat.index, y='popularity', label='Popularity', marker='o', color='#999999', linestyle='--')
# plt.title('Weighted Scores Comparison')
plt.xlabel('Gene Symbol', fontdict={'fontsize': 12})
plt.ylabel('Weighted Score',fontdict={'fontsize': 12})
plt.xticks(rotation=90)
plt.legend(loc='lower left')  # Adjust legend position and columns
plt.tight_layout()
# plt.savefig("./output/Figures/Fig1_weighted_scores_comparison_gpt4o.pdf", dpi=300)
plt.show()
```
